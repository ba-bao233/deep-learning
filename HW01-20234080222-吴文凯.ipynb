```python
import numpy as np

# 定义向量和矩阵
a = np.array([2, -1, 3])
b = np.array([1, 4, -2])
A = np.array([[1, 0, 2], [-1, 3, 1]])
B = np.array([[2, 1], [0, -1], [3, 2]])

# 1. 向量点积 a · b
dot_product = np.dot(a, b)
print("=" * 50)
print("1. 向量点积 a · b")
print("=" * 50)
print(f"a = {a}")
print(f"b = {b}")
print(f"a · b = {dot_product}")

# 2. 矩阵乘法 A × B
AB = np.dot(A, B)
print("\n" + "=" * 50)
print("2. 矩阵乘法 A × B")
print("=" * 50)
print(f"A = \n{A}")
print(f"\nB = \n{B}")
print(f"\nA × B = \n{AB}")
print(f"结果矩阵形状: {AB.shape}")

# 3. 向量 a 的 Frobenius 范数
frobenius_norm = np.linalg.norm(a)
print("\n" + "=" * 50)
print("3. 向量 a 的 Frobenius 范数")
print("=" * 50)
print(f"a = {a}")
print(f"||a||_F = {frobenius_norm:.4f}")
```

```python
# 3.1 贝叶斯公式

# 已知条件
P_D = 0.001          # 患病率 P(D)
P_T_pos_given_D = 0.99  # 灵敏度 P(T+|D)
P_T_pos_given_not_D = 0.02  # 假阳性率 P(T+|-D)

# 计算 P(T+) - 检测为阳性的总概率
P_T_pos = (P_T_pos_given_D * P_D) + (P_T_pos_given_not_D * (1 - P_D))

# 应用贝叶斯公式: P(D|T+) = P(T+|D) * P(D) / P(T+)
P_D_given_T_pos = (P_T_pos_given_D * P_D) / P_T_pos

print("=" * 50)
print("贝叶斯公式计算")
print("=" * 50)
print(f"P(D) = {P_D} (患病率)")
print(f"P(T+|D) = {P_T_pos_given_D} (灵敏度)")
print(f"P(T+|-D) = {P_T_pos_given_not_D} (假阳性率)")
print(f"\nP(T+) = P(T+|D)*P(D) + P(T+|-D)*P(-D)")
print(f"P(T+) = {P_T_pos_given_D}*{P_D} + {P_T_pos_given_not_D}*{1-P_D} = {P_T_pos:.6f}")
print(f"\nP(D|T+) = P(T+|D)*P(D) / P(T+)")
print(f"P(D|T+) = {P_T_pos_given_D}*{P_D} / {P_T_pos:.6f}")
print(f"\n检测阳性时真正患病的概率: {P_D_given_T_pos*100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 设置随机种子以便复现
np.random.seed(42)

# 参数设置
n = 10000  # 每次采样的样本数
m = 1000   # 重复次数

# 生成 m 组，每组 n 个 U(0,1) 随机变量，计算均值
means = [np.mean(np.random.uniform(0, 1, n)) for _ in range(m)]

# 理论值：U(0,1) 的均值=0.5，方差=1/12
# 样本均值的理论分布：N(0.5, 1/(12n))
theoretical_mean = 0.5
theoretical_var = 1 / (12 * n)
theoretical_std = np.sqrt(theoretical_var)

print("=" * 50)
print("中心极限定理验证")
print("=" * 50)
print(f"每次样本数 n = {n}")
print(f"重复次数 m = {m}")
print(f"\n理论值:")
print(f"  均值 = {theoretical_mean}")
print(f"  方差 = 1/(12*{n}) = {theoretical_var:.8f}")
print(f"  标准差 = {theoretical_std:.6f}")

print(f"\n实际模拟结果:")
print(f"  均值 = {np.mean(means):.6f}")
print(f"  方差 = {np.var(means):.8f}")
print(f"  标准差 = {np.std(means):.6f}")

print(f"\n相对误差:")
print(f"  均值误差 = {abs(np.mean(means) - theoretical_mean)/theoretical_mean * 100:.4f}%")
print(f"  方差误差 = {abs(np.var(means) - theoretical_var)/theoretical_var * 100:.4f}%")

# 绘制直方图 + 理论正态曲线
plt.figure(figsize=(10, 6))
plt.hist(means, bins=40, density=True, alpha=0.7, color='steelblue', 
         edgecolor='white', label='模拟均值分布')

x = np.linspace(0.45, 0.55, 100)
plt.plot(x, stats.norm.pdf(x, theoretical_mean, theoretical_std), 
         'r-', lw=2, label=f'理论正态 N({theoretical_mean}, {theoretical_var:.2e})')

plt.xlabel('样本均值', fontsize=12)
plt.ylabel('概率密度', fontsize=12)
plt.title(f'中心极限定理验证 (n={n}, m={m})', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import sympy as sp
import numpy as np

# 定义符号变量
w1, w2, x1, x2, y = sp.symbols('w1 w2 x1 x2 y')

# 已知值
x1_val, x2_val, y_val = 2, 1, 3
w1_val, w2_val = 0.5, 1

# 定义 z = (w1*x1 + w2*x2 - y)^2
u = w1*x1 + w2*x2 - y
z = u**2

print("=" * 50)
print("4.1 导数与反向传播")
print("=" * 50)
print(f"\n定义: z = (w1*x1 + w2*x2 - y)^2")
print(f"已知: x1={x1_val}, x2={x2_val}, y={y_val}")

# 1. 求偏导数表达式
dz_dw1 = sp.diff(z, w1)
dz_dw2 = sp.diff(z, w2)

print("\n" + "-" * 50)
print("1. 偏导数表达式:")
print("-" * 50)
print(f"∂z/∂w1 = {sp.simplify(dz_dw1)}")
print(f"∂z/∂w2 = {sp.simplify(dz_dw2)}")

# 2. 代入数值计算梯度
u_val = w1_val * x1_val + w2_val * x2_val - y_val
dz_dw1_val = float(dz_dw1.subs({w1: w1_val, w2: w2_val, x1: x1_val, x2: x2_val, y: y_val}))
dz_dw2_val = float(dz_dw2.subs({w1: w1_val, w2: w2_val, x1: x1_val, x2: x2_val, y: y_val}))

print("\n" + "-" * 50)
print(f"2. 代入 w1={w1_val}, w2={w2_val} 计算梯度:")
print("-" * 50)
print(f"u = w1*x1 + w2*x2 - y = {w1_val}*{x1_val} + {w2_val}*{x2_val} - {y_val} = {u_val}")
print(f"\n∂z/∂w1 = {dz_dw1_val}")
print(f"∂z/∂w2 = {dz_dw2_val}")
print(f"\n梯度向量: ∇z = [{dz_dw1_val}, {dz_dw2_val}]")

# 3. 使用数值方法验证
def z_func(w1, w2):
    return (w1*x1_val + w2*x2_val - y_val)**2

print("\n" + "-" * 50)
print("3. 数值梯度验证 (使用中心差分):")
print("-" * 50)
eps = 1e-5
numerical_dz_dw1 = (z_func(w1_val + eps, w2_val) - z_func(w1_val - eps, w2_val)) / (2 * eps)
numerical_dz_dw2 = (z_func(w1_val, w2_val + eps) - z_func(w1_val, w2_val - eps)) / (2 * eps)
print(f"数值 ∂z/∂w1 ≈ {numerical_dz_dw1:.6f}")
print(f"数值 ∂z/∂w2 ≈ {numerical_dz_dw2:.6f}")
print(f"\n解析解与数值解误差:")
print(f"  |∂z/∂w1 误差| = {abs(dz_dw1_val - numerical_dz_dw1):.2e}")
print(f"  |∂z/∂w2 误差| = {abs(dz_dw2_val - numerical_dz_dw2):.2e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 线性回归梯度下降
np.random.seed(42)

# 生成数据 y = 2x + 3 + noise
n = 100
X = 2 * np.random.rand(n, 1)
y = 3 + 2 * X + np.random.randn(n, 1)

# 添加偏置项
X_b = np.c_[np.ones((n, 1)), X]

# 梯度下降参数
lr = 0.01  # 学习率
epochs = 1000
m = n

# 初始化参数
theta = np.random.randn(2, 1)

# 梯度下降
loss_history = []
for epoch in range(epochs):
    gradients = (2/m) * X_b.T @ (X_b @ theta - y)
    theta = theta - lr * gradients
    loss = (1/m) * np.sum((X_b @ theta - y)**2)
    loss_history.append(loss)

print(f"学习到的参数: θ₀ = {theta[0][0]:.4f}, θ₁ = {theta[1][0]:.4f}")
print(f"真实参数: θ₀ = 3, θ₁ = 2")

# 可视化
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.scatter(X, y, alpha=0.5, label='数据点')
X_line = np.linspace(0, 2, 100).reshape(-1, 1)
X_line_b = np.c_[np.ones((100, 1)), X_line]
plt.plot(X_line, X_line_b @ theta, 'r-', lw=2, label='拟合直线')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('线性回归拟合')

plt.subplot(1, 2, 2)
plt.plot(loss_history)
plt.xlabel('迭代次数')
plt.ylabel('损失函数')
plt.title('梯度下降收敛')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy import stats

# 题目：某城市一天中气温 X ~ N(15, 5^2)，计算 P(X > 20)

# 已知参数
mu = 15   # 均值
sigma = 5  # 标准差

X = stats.norm(loc=mu, scale=sigma)

# 1. P(X > 20) - 一天气温大于20度的概率
prob_1_day = 1 - X.cdf(20)

# 2. P(至少有一天 > 20) - 三天中至少有一天大于20度
# 假设每天独立，P(至少一天>20) = 1 - P(三天都不>20)
prob_none = (X.cdf(20))**3
prob_at_least_1 = 1 - prob_none

print("=" * 50)
print("气温概率计算")
print("=" * 50)
print(f"气温分布: X ~ N({mu}, {sigma}^2)")
print()
print(f"1. 一天气温 > 20℃ 的概率:")
print(f"   P(X > 20) = 1 - Φ((20 - {mu})/{sigma})")
print(f"   P(X > 20) = 1 - Φ({(20-mu)/sigma:.4f})")
print(f"   P(X > 20) = {prob_1_day:.6f}")
print()
print(f"2. 三天中至少有一天 > 20℃ 的概率:")
print(f"   P(≥1天>20) = 1 - [P(X≤20)]^3")
print(f"   P(≥1天>20) = 1 - ({X.cdf(20):.6f})^3")
print(f"   P(≥1天>20) = {prob_at_least_1:.6f}")

# 3. 蒙特卡洛模拟验证
np.random.seed(42)
n_simulations = 100000

# 模拟每天的气温
simulated_days = np.random.normal(mu, sigma, (n_simulations, 3))

# 统计三天中至少一天>20的次数
days_above_20 = np.sum(simulated_days > 20, axis=1)
simulated_prob = np.sum(days_above_20 >= 1) / n_simulations

print()
print("3. 蒙特卡洛模拟验证:")
print(f"   模拟次数: {n_simulations}")
print(f"   模拟结果: P(≥1天>20) ≈ {simulated_prob:.6f}")
print(f"   理论值:   P(≥1天>20) = {prob_at_least_1:.6f}")
print(f"   相对误差: {abs(simulated_prob - prob_at_least_1)/prob_at_least_1 * 100:.4f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 蒙特卡洛模拟可视化
np.random.seed(42)

# 模拟参数
mu, sigma = 15, 5
n_simulations = 10000

# 模拟三天的气温
simulated_days = np.random.normal(mu, sigma, (n_simulations, 3))

# 统计每天>20的天数
days_above_20 = np.sum(simulated_days > 20, axis=1)

# 统计各情况出现次数
counts = [np.sum(days_above_20 == i) for i in range(4)]
probs = [c / n_simulations for c in counts]

print("三天中气温>20℃天数分布:")
print("-" * 30)
for i in range(4):
    print(f"  {i}天 > 20℃: 次数={counts[i]}, 概率={probs[i]:.4f}")

# 可视化
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(range(4), probs, color='steelblue', edgecolor='white')
plt.xlabel('Number of days > 20C')
plt.ylabel('Probability')
plt.title('Monte Carlo Simulation: Days Above 20C')
plt.xticks(range(4))
for i, p in enumerate(probs):
    plt.text(i, p + 0.01, f'{p:.3f}', ha='center')

plt.subplot(1, 2, 2)
# 绘制气温分布
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
from scipy import stats
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'b-', lw=2, label='N(15, 5^2)')
plt.axvline(x=20, color='r', linestyle='--', label='X=20')
plt.fill_between(x[x>20], stats.norm.pdf(x[x>20], mu, sigma), alpha=0.3, color='red', label='P(X>20)')
plt.xlabel('Temperature (C)')
plt.ylabel('Probability Density')
plt.title('Temperature Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```
